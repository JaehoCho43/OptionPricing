```python
import analytics.analyticsBS as analyticsBS
import analytics.analytics_utils as analytics_utils
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Parameters

rd = 0.03
rf = 0.04
T = 1
vol = 0.1
K = 1.1
spotL = 0.5
spotU = 1.5

numMeshX_PDE = 100
numMeshT_PDE = 100

X = np.linspace(spotL, spotU, 1000)
```

## EuroVanilla

```python
euroVanilla = analyticsBS.EuroVanilla(rd, rf, T, vol, K, analytics_utils.CallPut.CALL)

intrinsic_vec = np.vectorize(lambda S : euroVanilla.intrinsicValue(S))

intrinsicValues = intrinsic_vec(X)
priceExact = euroVanilla.priceExact(X)
priceMC = euroVanilla.priceMonteCarlo(X, numPaths = 50000)
X_PDE, pricePDE = euroVanilla.pricePDEwithSpots(spotL, spotU, numSpots = 2500, numT = 50)

plt.plot(X, intrinsicValues, label = "intrinsic")
plt.plot(X, priceExact, label = "exact")
plt.plot(X, priceMC, label = "Monte Carlo")
plt.plot(X_PDE, pricePDE, label = "PDE")
plt.title("euroVanilla")
plt.legend()
plt.plot()

fig, ax = plt.subplots(2,2)
ax[0,0].plot(X, intrinsicValues)
ax[0,0].set_title("intrinsic")
ax[0,1].plot(X, priceExact)
ax[0,1].set_title("exact")
ax[1,0].plot(X, priceMC)
ax[1,0].set_title("Monte Carlo")
ax[1,1].plot(X_PDE, pricePDE)
ax[1,1].set_title("PDE")
plt.tight_layout()
plt.plot()
```

## Binary

```python
binary = analyticsBS.Binary(rd, rf, T, vol, K, analytics_utils.CallPut.CALL)

intrinsic_vec = np.vectorize(lambda S : binary.intrinsicValue(S))

intrinsicValues = intrinsic_vec(X)
priceExact = binary.priceExact(X)
priceMC = binary.priceMonteCarlo(X, numPaths = 50000)
X_PDE, pricePDE = binary.pricePDEwithSpots(spotL, spotU, numSpots = 2500, numT = 50)
_, pricePDE_Rannacher = binary.pricePDEwithSpots(spotL, spotU, numSpots = 2500, numT = 50, numRanaccher = 1)

plt.plot(X, intrinsicValues, label = "intrinsic")
plt.plot(X, priceExact, label = "exact")
plt.plot(X, priceMC, label = "Monte Carlo")
plt.plot(X_PDE, pricePDE, label = "PDE")
plt.plot(X_PDE, pricePDE_Rannacher, label = "PDE with Rannacher")
plt.title("binary")
plt.legend()
plt.plot()

fig, ax = plt.subplots(2,3)
ax[0,0].plot(X, intrinsicValues)
ax[0,0].set_title("intrinsic")
ax[0,1].plot(X, priceExact)
ax[0,1].set_title("exact")
ax[0,2].plot(X, priceMC)
ax[0,2].set_title("Monte Carlo")
ax[1,0].plot(X_PDE, pricePDE)
ax[1,0].set_title("PDE")
ax[1,1].plot(X_PDE, pricePDE_Rannacher)
ax[1,1].set_title("PDE with Rannacher")
plt.tight_layout()
plt.plot()
```
